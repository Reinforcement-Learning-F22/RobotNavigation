To reduce the the state space dimensions, this notebook creates a smaller map with smaller robot

```python
# All plots will be displayed directly below the code cell that produced it
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
from PIL import Image
import cv2
```

```python
#scale parameters for both robot and map images
robo_scale = 0.05
map_scale = 0.2
```

## Smaller world

```python
# read robot image and scale it down
robo_im = Image.open("data/robot.png")
w, h = robo_im.size
robo_im = robo_im.resize((int(np.ceil(w*robo_scale)), int(np.ceil(h*robo_scale))))
robo_im.save("small_robot.png") # save the img
plt.imshow(robo_im)
```

## smaller world

```python
# read map image and scale it down
map_im = Image.open("data/map01.jpg")
w, h = map_im.size
map_im = map_im.resize((int(w*map_scale ), int(h*map_scale )))

map_im.save("small_map.jpg") # save the img
plt.imshow(map_im)
```

## Merge the two images

```python
def merge2(im1, im2, pose):
    """
    im1: map
    im2: robot
    """
    pose = tuple(int(i) for i in pose)
    w, h = im1.size
    im = Image.new("RGBA", (w, h))

    im.paste(im1)
    im.alpha_composite(im2, dest=pose)

    return im
```

```python
#put robot into center of new world map
w, h = map_im.size
map_robot_img = merge2(map_im, robo_im, (w/2, h/2))
plt.imshow(map_robot_img)
```
